```python
import pandas as pd
from pygeocoder import Geocoder
```

```python
station = pd.read_csv('station.csv')
station
```

```python
weather = pd.read_csv('weather.csv',parse_dates=['date'])
weather.sort_values(by='date',inplace=True)
weather.head(20)
```

```python
def get_zip_code(lat, lon):
    result = pd.Series()
    for i in range(len(list(lat))):
        result.set_value(label=i,value=Geocoder.reverse_geocode(lat[i], lon[i]).formatted_address.split(',')[2][4:9],takeable=False)
        
    return result
```

```python
station['zip_code'] = station.applymap(get_zip_code(station.lat, station.long))
```
